```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
dataset.isnull().any().any()
```

```python
df=dataset.drop('User ID',axis=1)
df
```

```python
df['Gender'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df.iloc[:,0]=lb.fit_transform(df.iloc[:,0])
df
```

```python
from scipy import stats
z=np.abs(stats.zscore(df))
z
```

```python
threshold=3
np.where(z>threshold)
```

```python
x=df.iloc[:,0:3].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#Normalization->Good practice(less variation between independent variables)
```

```python
#Normalization using standard scalar
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestClassifier()
```

```python
n_estimators=[int(x) for x in np.linspace(50,1000,12)]
max_depth=[int(x) for x in np.linspace(5,30,6)]
min_samples_leaf=[1,2,5,10]
criterion=['entropy','gini']
```

```python
random_grid = {'n_estimators':n_estimators,
                'max_depth':max_depth,
                'min_samples_leaf':min_samples_leaf,
                 'criterion':criterion
               }
random_grid
```

```python
rf_random= RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='accuracy', n_iter = 10, cv = 5,random_state=0)
```

```python
rf_random.fit(x_train,y_train)
```

```python
# https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
```

```python
rf_random.best_params_
```

```python
y_pred=rf_random.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
