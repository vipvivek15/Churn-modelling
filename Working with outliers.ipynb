# Z-score for Boston data

```python
from scipy import stats
```

```python
stats.zscore(df)
```

```python
dir(df)
```

```python
z=np.abs(stats.zscore(df))
z
```

```python
#define a threshold to identify an outlier
threshold=3
np.where(z>threshold)
```

```python
z[55][1]
```

```python
z[56][1]
```

```python
#Get clean data by removing outliers
```

```python
df_no_outlier=df[(z<=3).all(axis=1)]
df_no_outlier
```

```python
#Detecting outlier using IQR(Q3-Q1)
```

```python
Q1=df.quantile(0.25)
Q1
```

```python
Q3=df.quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
boston_iqr_outlier=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
boston_iqr_outlier
```
